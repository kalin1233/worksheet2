### Effective Programming

a) What is a drawback of the top down approach?

One drawback of the top-down approach is that it may lead to a delay in getting a working implementation of the system. Since the development starts from a high-level view and progresses towards lower levels of detail, it might take a significant amount of time before a functional prototype or version of the software is available. This delay can be problematic, especially in fast-paced development environments where stakeholders may need to see tangible progress sooner.

b) What is a drawback of the bottom up approach?

Bottom-up development focuses on individual components without a comprehensive view of the entire system. This can make it challenging to ensure that all components work seamlessly together to achieve the desired functionality.

Use logging effectively, performing unit tests, practice code review.

a) Complete the code below to read in a filepath to the `train.csv` and returns the DataFrame.

```python
import pandas as pd

df = pd.read_csv('train.csv')
df.describe()
```

b) Complete the code so it returns the number of rows that have at least one empty column value

```python
print("there are " +  str(df[df.isnull().any(axis=1)].shape[0]) + " rows with at least one empty value")
```

c) Complete the code below to remove all columns with more than 200 NaN values

```python
df = df.dropna(axis=1, thresh=len(df) - 200)
df.columns
```

d) Complete the code below to replaces `male` with 0 and `female` with 1

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df.head()
```

e) Complete the code below to add four columns `First Name`, `Middle Name`, `Last Name`, and `Title` corresponding to the value in the `name` column.

For example: `Braund, Mr. Owen Harris` would be: 

|First Name | Middle Name | Last Name | Title |
|-----------|-------------|-----------|-------|
| Owen      |  Harris     |  Braund   | Mr    |

Anything not clearly one of the above 4 categories can be ignored.

```python
df[['Last Name', 'Title', 'First Name', 'Middle Name']] = df['Name'].str.extract(r'^(.*),\s*(\w+)\.\s*(\w+)\s*(.*)$', expand=True)
df.head()
```

f) Complete the code below to replace all missing ages with the average age

```python
df['Age'] = df['Age'].mean()
df.head()
```

g) Plot a bar chart of the average age of those that survived and did not survive. Briefly comment on what you observe.

```python
survived_age = df[df['Survived'] == 1]['Age'].mean()
not_survived_age = df[df['Survived'] == 0]['Age'].mean()

plot_df = pd.DataFrame({
    'Survival Status': ['Survived', 'Not Survived'],
    'Average Age': [survived_age, not_survived_age]
})

ax = plot_df.plot(kind='bar', x='Survival Status', y='Average Age', legend=False)
ax.set_xlabel('Survival Status')
ax.set_ylabel('Average Age')
ax.set_title('Average Age of Survived and Not Survived Passengers')

```

The depicted chart illustrates that the average age for individuals who both survived and did not survive is approximately 30 years old.
